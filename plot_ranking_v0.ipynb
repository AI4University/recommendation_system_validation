### Validation Plot Ranking

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm

from funciones_recommendation_system import get_datasets, match_researcher_call, recommendation_system_researcher_call, match_call_researcher, recommendation_system_call_researcher
```

```python
# load databases
path = '/export/data_ml4ds/AI4U/Datasets/'
version_wp = '20240510'
version_rp = '20240321'

df_publications, df_projects, df_publications_researchers,df_projects_researchers, df_researchers, df_calls = get_datasets(path, version_wp, version_rp)

# load the validation set
df_val = pd.read_excel('/export/usuarios_ml4ds/mafuello/Github/recommendation_system_validation/validation_set.xlsx')
df_val = df_val.rename(columns={'research portal': 'invID'})
df_val = df_val.rename(columns={'invID': 'id_RP'})#.merge(df_researchers[['id_RP', 'id_researcher']], on='id_RP')
```

### RECOMENDATIONS CALLS

```python
methods = ['BERT', 'bhattacharyya', 'separated', 'semiseparated']
agg_methods = ['sum', 'mean', 'mean_imp'] 

method = methods[3]

dict_posiciones_sum = {i: 0 for i in range(1, 767)}
dict_posiciones_mean = {i: 0 for i in range(1, 767)}
dict_posiciones_mean_imp = {i: 0 for i in range(1, 767)}


for i in df_val.index:
    invID = df_val['id_researcher'][i]
    
    if method == 'semiseparated':
        invID = df_val['id_researcher'][i].astype(str) # SOLO VALE PARA semiseparated
    
    call = df_val['LÃ­nea prioritia/panel/topic'][i]   
    try:
        #SUM
        ranking_sum = recommendation_system_researcher_call(method=method, agg_method='sum', researcher=invID, calls=df_calls)
        indice_valor_exacto_sum = ranking_sum.loc[ranking_sum['Call'] == call].index[0] + 1
        dict_posiciones_sum[indice_valor_exacto_sum] += 1
        #MEAN
        ranking_mean = recommendation_system_researcher_call(method=method, agg_method='mean', researcher=invID, calls=df_calls)
        indice_valor_exacto_mean = ranking_mean.loc[ranking_mean['Call'] == call].index[0] + 1
        dict_posiciones_mean[indice_valor_exacto_mean] += 1
        #MEAN IMP
        ranking_mean_imp = recommendation_system_researcher_call(method=method, agg_method='mean_imp', researcher=invID, calls=df_calls)
        indice_valor_exacto_mean_imp = ranking_mean_imp.loc[ranking_mean_imp['Call'] == call].index[0] + 1
        dict_posiciones_mean_imp[indice_valor_exacto_mean_imp] += 1
        
    except:
        print(i)
```

```python
# Extraer las claves y los valores de cada diccionario
keys1, values1 = list(dict_posiciones_sum.keys()), list(dict_posiciones_sum.values())
keys2, values2 = list(dict_posiciones_mean.keys()), list(dict_posiciones_mean.values())
keys3, values3 = list(dict_posiciones_mean_imp.keys()), list(dict_posiciones_mean_imp.values())

# Crear el grÃ¡fico
plt.figure(figsize=(10, 5))

# Plotear cada diccionario
plt.plot(keys1, values1, marker='o', linestyle='-', color='blue', label='Data1')
plt.plot(keys2, values2, marker='s', linestyle='-', color='red', label='Data2')
plt.plot(keys3, values3, marker='^', linestyle='-', color='green', label='Data3')

# AÃ±adir tÃ­tulos y etiquetas
plt.title('Plot de Tres Diccionarios')
plt.xlabel('Keys')
plt.ylabel('Values')

# AÃ±adir leyenda
plt.legend()

# Mostrar el grÃ¡fico
plt.grid(True)
plt.show()
```
